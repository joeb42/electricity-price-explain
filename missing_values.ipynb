```python
from pathlib import Path
import matplotlib.pyplot as plt
plt.style.use("seaborn")
import pandas as pd
import seaborn as sns

data_path = Path("data/")

X = pd.read_csv(next(data_path.glob("X_train*")))
y = pd.read_csv(next(data_path.glob("y*")))
df = pd.merge(X, y, on="ID")
df
```

```python
(df.isna()
    .mean()
    .plot(kind="bar")
)
```

```python
df.loc[:, df.isna().any()]
```

```python
from sklearn.impute import SimpleImputer 

df_imputed = df.copy()
imputer = SimpleImputer()
df_imputed.loc[:, df.isna().any()] = imputer.fit_transform(df.loc[:, df.isna().any()] )
df_imputed
```

```python
fig, ax = plt.subplots(1, 2, figsize=(28, 10))
sns.heatmap(df.corr("spearman"), ax=ax[0])
sns.heatmap(df_imputed.corr("spearman"), ax=ax[1])
```

```python
sns.heatmap(df.corr("spearman") - df_imputed.corr("spearman"))
```
